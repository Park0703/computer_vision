## Composite pattern (Composition)

```
상속을 하지 않고 클래스내에 객체를 불러와 다른 클래스의 일부 기능을 사용하는 방법.
composite pattern의 활용 중에서 상속을 대신하는 기능이 있다 
```

`````python
class A:
  pass

class B:
  def __init__(self):
    self.a = A()
    self.c = B()

  def multi(self):
    return self.a * self.b 
`````

`````python
class AA:
  a = 1 
  def aa(self, x):
    self.b = x
  
  def aaa(self):
    return self.b 

class BB(AA):  # AA와 coupling (밀접하게 연관 되어 있다/AA가 바뀌면 BB도 같이 바뀌어 버린다)
  pass
`````

`````python
class CC:
  a = 1 
  def __init__(self):
    self.y = AA() 
  def aa(self, x):
    self.y.b = x 
  def aaa(self):
    return self.y.b
`````

`````python
a = AA()
`````

`````python
a.aa(4) # 인스턴스 변수 b가 4로 바뀜 
`````

`````python
a.aaa()
`````

`````python
c = CC()
`````

`````python
c.aa(3) # 인스턴스 변수가 
`````

`````python
c.aaa()
`````

### Python의 inheritance는 delgate이다 

`````python
class DD: ## AA가 변할때 같이 변함 
  a = 2 # overriding
  def __init__(self):
    self.y = AA()
  def aa(self, x):
    self.y.aa(x)
  def aaa(self):
    return self.y.aaa()
`````

`````python
class EE:
  def __init__(self):
    self.y = AA()
  def aa(self): # overriding함으로써 함수 aa의 기능을 바꾸어 버렸다 
    print('aa')
  def __getattr__(self, x): # 인스턴스 변수가 없을 때 attributeError가 발생한다 
    return getattr(self.y, x)  
`````

## Descriptor

 
```python
점(.)으로 객체의 멤버를 접근할 때, 먼저 인스턴스 변수(dict)에서 멤버를 찾는다. 없을 경우 클래스 변수에서 찾는다. 
클래스에서 멤버를 찾고 객체가 descriptor 프로토콜을 구현했다면 바로 멤버를 리턴하지 않고 descriptor 메소드(get, set, delete)를 호출한다

1. setter  # 값을 지정하고 
2. getter  # 값을 조회하고 
3. deleter # 값을 지우고
```

### Descriptor를 만드는 3가지 방법 


```python
1. composition
2. Property
3. Decorator
```

### Composition 방법

`````python
class Name: # descriptor
  def __init__(self):
    print('__init__')
    self._name = ''
  def __get__(self, instance, owner): # 관례상 owner라고 함 
    print('__get__')
    return self._name
  def __set__(self, instance, name):
    print('__set__')
    self._name = name
  def __delete__(self, instance):
    print('__delete__')
    del self._name
`````

`````python
class Person:
  name = Name()
`````

`````python
p = Person()
`````

`````python
p.name
`````

`````python
p.name = 'sun'
`````

`````python
del p.name
`````

### Property 방법

`````python
class Person2: 
  def __init__(self):
    print('__init__')
    self._name = ''
  def fget(self):
    print('__get__')
    return self._name
  def fset(self, name):
    print('__set__')
    self._name = name
  def fdelete(self):
    print('__delete__')
    del self._name
  name = property(fget, fset, fdelete)
`````

`````python
p2 = Person2()
`````

`````python
p2.name
`````

`````python
p2.fget()
`````

`````python
p2.name = 'name'
`````

`````python
p2.name
`````

`````python
p2.fset('ff')
`````

`````python
p2.name
`````

`````python
p2.fdelete()
`````

### Decorator 방법

`````python
class Person3:
  def __init__(self):
    self._name = ''
  
  @property
  def name(self):
    print('__get__')
    return self._name
  

  @name.setter
  def name(self, name):
    print('__set__')
    self._name = name 
  
  @name.deleter
  def name(self):
    print('__delete__')
    del self._name
`````

`````python
p3 = Person3()
`````

`````python
p3.name
`````

`````python
p3.name = '123'
`````

`````python
p3.name 
`````

`````python
del(p3.name)
`````

### Singletone

인스턴스를 하나만 만들 수 있는 클래스

`````python
class Singleton(type):
    _instance = None
    def __call__(cls, *args, **kwarg): # class에서 괄호 붙일 떄 어떻게 사용하는지 정의 한다 
        if not cls._instance:
             cls._instance = super(Singleton, cls).__call__(*args, **kwarg)      
	                    # super().__call__(*args, **kw) 동일
        return cls._instance    
`````

`````python
class MyClass(metaclass=Singleton):
  pass
`````

`````python
m = MyClass()
s = MyClass()
`````

`````python
id(m) == id(s)
`````

`````python
class A(object, metaclass=type): # object는 type이라는 클래스로부터 정의된 가장 핵심적인 클래스 
  pass
`````

## Operator

### 모든 연산자는 그에 대응하는 dundu(magic/special)가 있다 

`````python
class B:
    def __init__(self, num):
        self.num = num 
 
    # 연산자 +에 대응하는 dundu method 
    def __add__(self, x, y):
        return x + y
 
    # 연산자 -에 대응하는 dundu method 
    def __sub__(self, a): 
        self.a -= a
 
    # 연산자 or에 대응하는 dundu method     
    def __or__(self, num):
        self.Num = num|Num     
`````
